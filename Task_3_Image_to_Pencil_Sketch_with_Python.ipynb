Importing libraries

```python
import cv2
from google.colab.patches import cv2_imshow
```

Reading image

```python
image=cv2.imread('/content/drive/MyDrive/Lets Grow More Tasks/task3_image.jpg')
cv2_imshow(image)
```

Converting RBG image to grayscale image

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

Inverting grayscale image (Negative image)

```python
inverted_image = 255 - gray_image
cv2_imshow(inverted_image)
```

Blurring Inverted image

```python
blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), 0)
cv2_imshow(blurred_image)
```

Mixing grayscale image with blurry inverted image in order to get our pencil sketch

```python
inverted_blurred = 255 - blurred_image
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2_imshow(pencil_sketch)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 7))

fig.add_subplot(1,2,1)
plt.imshow(image)
plt.axis('off')
plt.title('Original Image')

fig.add_subplot(1,2,2)
plt.imshow(pencil_sketch,cmap="gray")
plt.axis('off')
plt.title('Pencil Sketch')
```
